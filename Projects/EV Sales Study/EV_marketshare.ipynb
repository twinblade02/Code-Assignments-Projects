```python
# World analysis, car sales share
vis1 = raw.query('region == "World" & category == "Historical" & parameter == "EV sales share" & mode == "Cars"')
vis2 = raw.query('region == "World" & category == "Historical" & parameter == "EV charging points"')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

ax[0].plot(vis1['year'], vis1['value'])

# Plot the second line with hue

vis2.pivot(index='year', columns='powertrain', values='value').plot(ax=ax[1])

plt.tight_layout()
plt.show()
```
