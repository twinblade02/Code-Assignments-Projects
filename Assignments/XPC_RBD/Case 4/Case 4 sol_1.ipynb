```python
# 6. unique tokens including stopwords
stop = set(stopwords.words('english'))
dump_stop = []
for r in data_copy_df['Text']:
    word_token = word_tokenize(r)
    filtered = [w for w in word_token if not w.lower() in stop]
    filtered = []
    for w in word_token:
        if w not in stop:
            filtered.append(w)
    dump_stop.append(filtered)
    #print(word_token)
    #word_array = np.array(word_token)
    #np.unique(word_array, return_counts=True)

print(len(set([x[0] for x in dump_stop])))
    
'''
Process - tokenize words in text field, then filter stopwords out. Toss filtered sentences into a 
'dump' variable to count unique tokens by leveraging the set keyword
'''
```

```python
# counting frequencies
def count_these():
    freq = {}
    for i in negative['tokenized_text']:
        for x in i:
            if x in freq:
                freq[x] += 1
            else:
                freq[x] = 1
    return freq
```

```python
# get sorted frequencies 
for w in sorted(freq, key=itemgetter(0), reverse = True):
    print(w, freq[w])
```
